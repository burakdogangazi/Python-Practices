```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.arange(1,6)
```

```python
x
```

```python
y = np.arange(2,11,2)
```

```python
plt.plot(x,y,"red")
plt.show()
```

```python
y
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,"blue")
#ikiye iki grafikte ilk olan
plt.subplot(2,2,2)
plt.plot(y,x,"yellow")
plt.subplot(2,2,3)
plt.plot(y,x,"red")
plt.subplot(2,2,4)
plt.plot(x,x**2,"black")
plt.show()
```

```python
fig,axes = plt.subplots(nrows = 2,ncols = 1)

#for ax in axes:
#    ax.plot(x,y)
 
axes[0].plot(x,y)
axes[0].set_title("First Axes")
axes[1].plot(x,x**3)
axes[1].set_title("Second Axes")

    

plt.tight_layout()
```

```python
fig = plt.figure(figsize = (8,6))

axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,color = "purple")
```

```python
fig,axes = plt.subplots(nrows = 2,ncols = 1,figsize = (8,6))
axes[0].plot(x,y,color = "red")
axes[0].set_title("First Axes")

axes[1].plot(x,x**0.5,color = "purple")
axes[1].set_title("Second Axes")

plt.tight_layout()
```

```python
fig = plt.figure(figsize = (6,4))

axes = fig.add_axes([0,0,1,1])

axes.plot(x,x ** 0.5,color = "red",label = "x karekök")

axes.plot(x,x**2, color = "yellow",label = "x kare")

axes.plot(x,x**3, color  = "#4055c2",label = "x küp")

axes.legend()
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
axes.plot(x,x**2,color = "red",linewidth = 5,linestyle = "--",marker = "o",markersize = 10,markerfacecolor = "black",markeredgecolor = "yellow",markeredgewidth = 5)

axes.set_xlim(0,15)
axes.set_ylim(0,50)
```
